# Arxiv 논문 데이터 수집

```python
import requests
import feedparser

# arXiv API endpoint
base_url = "http://export.arxiv.org/api/query"

# 검색 파라미터 설정

#   - cs.AI (인공지능)
#   - cs.LG (머신러닝)
#   - cs.CL (자연어처리, LLM 관련)
#   - cs.IR (정보검색, RAG 관련)
#   - cs.CV (컴퓨터 비전)
#   - cs.SD (소프트웨어 개발)
#   - cs.SE (소프트웨어 엔지니어링)
#   - cs.PL (프로그래밍 언어)
#   - cs.DB (데이터베이스)
#   - cs.DC (데이터 압축)
#   - cs.DM (데이터 마이닝)
#   - stat.ML (통계적 머신러닝)

search_query = "cat:cs.AI"  # AI 관련 논문
start = 0                   # 시작 인덱스
max_results = 50            # 가져올 논문 수

# API 요청
params = {
    "search_query": search_query,
    "start": start,
    "max_results": max_results
}
response = requests.get(base_url, params=params)
feed = feedparser.parse(response.text)

# 논문 목록 출력
for entry in feed.entries:
    print(f"Title: {entry.title}")
    print(f"Authors: {', '.join(author.name for author in entry.authors)}")
    print(f"Published: {entry.published}")
    print(f"Summary: {entry.summary[:200]}...")  # 요약 일부만 출력
    print(f"Link: {entry.link}")
    print("-" * 80)
```
